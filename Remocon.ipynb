```python
import clr
import os
import sys
import time
from System import Decimal, Int32

# 경로 설정
kinesis_path = r"C:\Program Files\Thorlabs\Kinesis"
sys.path.append(kinesis_path)
os.environ['PATH'] += ';' + kinesis_path

# .NET 어셈블리 참조
clr.AddReference("Thorlabs.MotionControl.DeviceManagerCLI")
clr.AddReference("Thorlabs.MotionControl.Benchtop.StepperMotorCLI")

# 네임스페이스 import
from Thorlabs.MotionControl.DeviceManagerCLI import DeviceManagerCLI
from Thorlabs.MotionControl.Benchtop.StepperMotorCLI import BenchtopStepperMotor

# 디바이스 목록 새로고침
DeviceManagerCLI.BuildDeviceList()
time.sleep(0.5)

# 시리얼 번호
serial = "70203624"

# 장비 객체 생성 및 연결
stepper = BenchtopStepperMotor.CreateBenchtopStepperMotor(serial)
stepper.Connect(serial)

# 채널 가져오기
channel = stepper.GetChannel(1)

# 1. Polling 시작
channel.StartPolling(250)
time.sleep(0.5)

# 2. 설정 초기화 기다리기
if not channel.IsSettingsInitialized():
    print("⚙️ 설정 초기화 중...")
    channel.WaitForSettingsInitialized(5000)

print("✅ 장비 설정 초기화 완료됨")

# ✅ 모터 설정 강제 로딩
channel.LoadMotorConfiguration(serial)
print("🔧 모터 구성 로드 완료")

# 3. 홈 위치로 이동
channel.Home(60000)
print("🏠 홈 이동 완료")

# 4. 원하는 위치로 이동
channel.MoveTo(Decimal(50000), Int32(60000))
print("🚀 이동 완료!")
```

```python
import clr
from System import Decimal, Int32
import time

# Thorlabs DLL 경로가 PATH에 있거나 pythonnet에서 로드 가능한 상태여야 함
clr.AddReference("Thorlabs.MotionControl.GenericMotorCLI")  # 모터 제어용 DLL (이름 확인 필요)

from Thorlabs.MotionControl.GenericMotorCLI import MotorDeviceFactory

serial_number = "70203624"

# 1. 장치 생성
device = MotorDeviceFactory.CreateMotorDevice(serial_number)
if device is None:
    raise Exception(f"장치 생성 실패: {serial_number}")

# 2. 장치 연결
device.Connect()
time.sleep(1)  # 연결 안정화 대기

# 3. 모터 채널 얻기 (채널 번호 1부터 시작)
channel = device.GetMotorChannel(1)

# 4. 모터 설정 불러오기
channel.LoadMotorConfiguration(serial_number)

# 5. 모터 초기화 (홈 위치 확인)
channel.Initialize()

# 6. 모터 홈 위치로 이동
if not channel.IsHomed:
    channel.Home(60000)  # 타임아웃 60초
    print("홈 위치로 이동 완료")

# 7. 모터 목표 위치로 이동 (예: 50000 steps)
channel.MoveTo(Decimal(50000), Int32(60000))  # 60초 타임아웃
print("목표 위치로 이동 완료")

# 8. 연결 종료
device.Disconnect()
```

```python
import Thorlabs.MotionControl.DeviceManagerCLI as DeviceManagerCLI

candidate_methods = [
    'BuildDeviceList', 'ScanEthernetRange', 'IsDeviceConnected', 'GetDeviceTypesList',
    'GetManualDeviceEntries', 'GetManualDeviceEntriesSize'
]

for method in candidate_methods:
    if hasattr(DeviceManagerCLI, method):
        print(f"Calling {method}()")
        func = getattr(DeviceManagerCLI, method)
        try:
            result = func()
            print(f"{method} result: {result}")
        except Exception as e:
            print(f"{method} 호출 실패: {e}")
    else:
        print(f"{method} 없음")
```

```python
import Thorlabs.MotionControl.DeviceManagerCLI as DeviceManagerCLI

factory = DeviceManagerCLI.DeviceFactory

methods = [m for m in dir(factory) if not m.startswith('_')]

print("DeviceFactory methods:")
print(methods)

for m in methods:
    attr = getattr(factory, m)
    if callable(attr):
        try:
            result = attr()
            print(f"{m}() 호출 성공, 결과: {result}")
        except Exception as e:
            print(f"{m}() 호출 실패: {e}")
    else:
        print(f"{m} : {attr}")

        
```
