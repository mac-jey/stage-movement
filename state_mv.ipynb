# Thorlabs Stage Movement
Example code to move a stage using the Kinesis SDK.

```python
import clr
import os
import sys

# Update this path to the Kinesis installation
kinesis_path = r'C:\Program Files\Thorlabs\Kinesis'
sys.path.append(kinesis_path)
os.environ['PATH'] += ';' + kinesis_path

clr.AddReference(os.path.join(kinesis_path, 'Thorlabs.MotionControl.DeviceManagerCLI.dll'))
clr.AddReference(os.path.join(kinesis_path, 'Thorlabs.MotionControl.KCube.DCServoCLI.dll'))

from Thorlabs.MotionControl.DeviceManagerCLI import DeviceManagerCLI
from Thorlabs.MotionControl.KCube.DCServoCLI import KCubeDCServo
```

```python
DeviceManagerCLI.BuildDeviceList()
serial_no = 'YOUR_SERIAL_NO'

motor = KCubeDCServo.CreateKCubeDCServo(serial_no)
motor.Connect(serial_no)

if not motor.IsConnected:
    raise RuntimeError('Failed to connect to device')

motor.WaitForSettingsInitialized(5000)
motor.StartPolling(250)
motor.EnableDevice()
motor.Home(60000)
motor.WaitForMoveToComplete(60000)
print('Homing complete')
```

```python
start_pos = motor.Position
print(f'Current position: {start_pos} mm')

target = start_pos + 1.0  # move 1 mm forward
motor.MoveTo(target, 60000)
motor.WaitForMoveToComplete(60000)

end_pos = motor.Position
print(f'Moved to: {end_pos} mm')
```

```python
motor.StopPolling()
motor.Disconnect()
print('Device disconnected')
```
